#IMPORTACION DE DATOS#

```python
import pandas as pd

url1 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv'
url2 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv'
url3 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv'
url4 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv'

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()

```

#1. Análisis de facturación

```python
#Sumar ventas de cada tienda
ingreso_tienda1=tienda1['Precio'].sum()
ingreso_tienda2=tienda2['Precio'].sum()
ingreso_tienda3=tienda3['Precio'].sum()
ingreso_tienda4=tienda4['Precio'].sum()

#Imprimir resultados
print(f"Ingreso Total Tienda 1: ${ingreso_tienda1:.2f}")
print(f"Ingreso Total Tienda 2: ${ingreso_tienda2:.2f}")
print(f"Ingreso Total Tienda 3: ${ingreso_tienda3:.2f}")
print(f"Ingreso Total Tienda 4: ${ingreso_tienda4:.2f}")

#Ingreso de todas las tiendas
ingreso_total=ingreso_tienda1+ingreso_tienda2+ingreso_tienda3+ingreso_tienda4
print(f"Ingreso Total de las 4 tiendas: ${ingreso_total:.2f}")
```

# 2. Ventas por categoría

```python

#Imprimiendo resultados por cada tienda
print(f"\nIngreso Total Tienda 1: {tienda1.groupby('Categoría del Producto')['Precio'].sum()}")
print(f"\nIngreso Total Tienda 2: {tienda2.groupby('Categoría del Producto')['Precio'].sum()}")
print(f"\nIngreso Total Tienda 3: {tienda3.groupby('Categoría del Producto')['Precio'].sum()}")
print(f"\nIngreso Total Tienda 4: {tienda4.groupby('Categoría del Producto')['Precio'].sum()}")
```

# 3. Calificación promedio de la tienda

```python
print(f"\nClasificación promedio en Tienda 1: {tienda1['Calificación'].mean():.2f}")
print(f"\nClasificación promedio en Tienda 2: {tienda2['Calificación'].mean():.2f}")
print(f"\nClasificación promedio en Tienda 3: {tienda3['Calificación'].mean():.2f}")
print(f"\nClasificación promedio en Tienda 4: {tienda4['Calificación'].mean():.2f}")
```

# 4. Productos más y menos vendidos

```python
print(f"\nLos productos mas vendidos en Tienda 1 son:")
display(tienda1.groupby("Producto")["Precio"].sum().sort_values(ascending=False).reset_index().head(3))

print(f"\nLos productos mas vendidos en Tienda 2 son:")
display(tienda2.groupby("Producto")["Precio"].sum().sort_values(ascending=False).reset_index().head(3))

print(f"\nLos productos mas vendidos en Tienda 3 son:")
display(tienda3.groupby("Producto")["Precio"].sum().sort_values(ascending=False).reset_index().head(3))

print(f"\nLos productos mas vendidos en Tienda 4 son:")
display(tienda4.groupby("Producto")["Precio"].sum().sort_values(ascending=False).reset_index().head(3))
```

```python
print(f"\nLos productos menos vendidos en Tienda 1 son:")
display(tienda1.groupby("Producto")["Precio"].sum().sort_values(ascending=True).reset_index().head(3))

print(f"\nLos productos menos vendidos en Tienda 2 son:")
display(tienda2.groupby("Producto")["Precio"].sum().sort_values(ascending=True).reset_index().head(3))

print(f"\nLos productos menos vendidos en Tienda 3 son:")
display(tienda3.groupby("Producto")["Precio"].sum().sort_values(ascending=True).reset_index().head(3))

print(f"\nLos productos menos vendidos en Tienda 4 son:")
display(tienda4.groupby("Producto")["Precio"].sum().sort_values(ascending=True).reset_index().head(3))
```

#5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
print(f"\nCosto de envio promedio en Tienda 1: {tienda1['Costo de envío'].mean():.2f}")
print(f"\nCosto de envio promedio en Tienda 2: {tienda2['Costo de envío'].mean():.2f}")

print(f"\nCosto de envio promedio en Tienda 3: {tienda3['Costo de envío'].mean():.2f}")
print(f"\nCosto de envio promedio en Tienda 4: {tienda4['Costo de envío'].mean():.2f}")
```

#GENERANDO GRAFICOS

```python
#Uniendo Tiendas en un solo DataFrame
tienda1['Tienda']=1
tienda2['Tienda']=2
tienda3['Tienda']=3
tienda4['Tienda']=4
df_total=pd.concat([tienda1,tienda2,tienda3,tienda4])
df_total.head()
```

```python
import matplotlib.pyplot as plt

df_total['Precio'] = df_total['Precio'].astype(float)

# Asegurar tipo correcto
df_total['Precio'] = df_total['Precio'].astype(float)

# El ingreso de cada venta es simplemente el precio
df_total['Ingreso'] = df_total['Precio']

# Agrupar datos
ingresos_totales = df_total.groupby('Tienda')['Ingreso'].sum()
ingreso_promedio = df_total.groupby('Tienda')['Ingreso'].mean()
distribucion_ingresos = ingresos_totales

colores = ['royalblue','orange','forestgreen','tomato']
```

```python


# 1. Barra vertical: Total de ingresos
plt.figure(figsize=(8,5))
ingresos_totales.plot(kind='bar',color=colores)
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Total Ingresos (USD)')
plt.xlabel('Tienda')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# 2. Torta/Pie: Distribución del total
plt.figure(figsize=(7,7))
distribucion_ingresos.plot(kind='pie', autopct='%1.2f%%', startangle=90)
plt.title('Participación de Cada Tienda en el Total de Ingresos')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# 3. Línea: Promedio de ingreso por venta
plt.figure(figsize=(8,5))

# Calculate average revenue per store using the combined dataframe
ingreso_promedio = df_total.groupby('Tienda')['Precio'].mean()

ingreso_promedio.plot(kind='line', marker='x', linestyle='--', color='orange')
plt.title('Ingreso Promedio de Venta por Tienda')
plt.ylabel('Ingreso Promedio (USD)')
plt.xlabel('Tienda')
plt.xticks(ingreso_promedio.index)  # Set x-axis ticks to store numbers
plt.grid(True)
plt.tight_layout()
plt.show()
```
